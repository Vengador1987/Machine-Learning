```python
enlaces = ["https://bigbangtrans.wordpress.com/series-1-episode-1-pilot-episode/",
           "https://bigbangtrans.wordpress.com/series-1-episode-2-the-big-bran-hypothesis/",
           "https://bigbangtrans.wordpress.com/series-1-episode-3-the-fuzzy-boots-corollary/",
           "https://bigbangtrans.wordpress.com/series-1-episode-4-the-luminous-fish-effect/",
           "https://bigbangtrans.wordpress.com/series-1-episode-5-the-hamburger-postulate/"]
```

```python
#Entramos en las URLS, y sacamos el texto

textos = list()

for enlace in enlaces:
    sleep(3)
    
    print(enlace)
    
    response = requests.get(enlace)
    
    soup = BeautifulSoup(response.text, "html.parser")
    
    texto = soup.find("div", class_ = "widecolumn").find_all("p")
    
    texto = [t.text for t in texto]
    
    textos.append(texto)
```

```python
dialogos = list()

for texto in textos:
    for oracion in texto:
        dialogos.append(oracion)
    
dialogos
```

```python
leonar = list()

for oracion in dialogos:
    if oracion.split()[0] == "Leonard:":
        leonar.append(oracion)
        
        
leonar
```

```python
personajes = ["Leonard:", "Sheldon:", "Penny:", "Raj:", "Howard:"] 
```

```python
def obtener_personajes(personaje):
    lista_personaje = list()

    for oracion in dialogos:
        try:
            if oracion.split()[0] == personaje:
                lista_personaje.append(oracion)
        except:
            continue

    return(lista_personaje)
```

```python
Leonard = obtener_personajes("Leonard:")
Sheldon = obtener_personajes("Sheldon:")
Penny = obtener_personajes("Penny:")
Raj = obtener_personajes("Raj:")
Howard = obtener_personajes("Howard:")
```

```python
tokens_leonard = nltk.word_tokenize(text = " ".join(Leonard).lower(), language = "English")
tokens_sheldon = nltk.word_tokenize(text = " ".join(Sheldon).lower(), language = "English")
tokens_penny = nltk.word_tokenize(text = " ".join(Penny).lower(), language = "English")
tokens_raj = nltk.word_tokenize(text = " ".join(Raj).lower(), language = "English")
tokens_howard = nltk.word_tokenize(text = " ".join(Howard).lower(), language = "English")
    
```

```python
def limpieza_stopword(lista_tokens):
    
    texto_limpio = list()
    
    stopwords = nltk.corpus.stopwords.words("English")

    
    for token in lista_tokens: 
        if token not in stopwords: 
            if len(token) > 1: 
                texto_limpio.append(token)
            
    return(texto_limpio)
```

```python
tokens_limpios_leonard = limpieza_stopword(tokens_leonard)
tokens_limpios_sheldon = limpieza_stopword(tokens_sheldon)
tokens_limpios_penny   = limpieza_stopword(tokens_penny)
tokens_limpios_raj     = limpieza_stopword(tokens_raj)
tokens_limpios_howard  = limpieza_stopword(tokens_howard)
```

```python
def riqueza_lexica_fun(tokens_limpios):
    
    total_palabras = len(tokens_limpios)
    palabras_diferentes = len(set(tokens_limpios))
    
    riqueza_lexica = palabras_diferentes / total_palabras
    
    return riqueza_lexica

   
```

```python
print(f' riqueza lexia Leonar: {riqueza_lexica_fun(tokens_limpios_leonard)}')
print(f' riqueza lexia Sheldon: {riqueza_lexica_fun(tokens_limpios_sheldon)}')
print(f' riqueza lexia Penny: {riqueza_lexica_fun(tokens_limpios_penny)}')
print(f' riqueza lexia Raj: {riqueza_lexica_fun(tokens_limpios_raj)}')
print(f' riqueza lexia Howard: {riqueza_lexica_fun(tokens_limpios_howard)}')
```

```python
texto_nltk_leonard = nltk.Text(tokens = tokens_limpios_leonard) 
```

```python
def latiguillo(tokens_limpios):
    
    texto_nltk = nltk.Text(tokens = tokens_limpios) 
    
    distribucion = nltk.FreqDist(samples = texto_nltk)

    return distribucion.most_common(4)[1:]
```

```python
print(f' Tres palabaras mas repetidas de Leonar: {latiguillo(tokens_limpios_leonard)}')
print(f' Tres palabaras mas repetidas de Sheldon: {latiguillo(tokens_limpios_sheldon)}')
print(f' Tres palabaras mas repetidas de Penny: {latiguillo(tokens_limpios_penny)}')
print(f' Tres palabaras mas repetidas de Raj: {latiguillo(tokens_limpios_raj)}')
print(f' Tres palabaras mas repetidas de Howard: {latiguillo(tokens_limpios_howard)}')
```

```python
def palabaras_similares(tokens_limpios):

    similares = list()

    for word in nltk.word_tokenize(latiguillo(tokens_limpios)[0][0]):
        similares.append(nltk.text.ContextIndex(tokens_limpios).similar_words(word))

    return pd.DataFrame(data = similares, index = [nltk.word_tokenize(latiguillo(tokens_limpios)[0][0])]).T
    
```

```python
def hapax_personaje(tokens_limpios):
    # .Text()
    texto_nltk = nltk.Text(tokens = tokens_limpios)

    # .FreqDist()
    distribucion = nltk.FreqDist(samples = texto_nltk)

    # .hapaxes
    hapaxes = distribucion.hapaxes()

    return hapaxes
```

```python
print(f' hapaxes de Leonar: {hapax_personaje(tokens_limpios_leonard)}')
print(f' hapaxes de Sheldon: {hapax_personaje(tokens_limpios_sheldon)}')
print(f' hapaxes de Penny: {hapax_personaje(tokens_limpios_penny)}')
print(f' hapaxes de Raj: {hapax_personaje(tokens_limpios_raj)}')
print(f' hapaxes de Howard: {hapax_personaje(tokens_limpios_howard)}')
```

```python
def Grafica_20_palabras(tokens_limpios):

    nltk.FreqDist(nltk.Text(tokens_limpios)).plot(20)

    plt.show()
```

```python
Grafica_20_palabras(tokens_limpios_penny)
Grafica_20_palabras(tokens_limpios_sheldon)
Grafica_20_palabras(tokens_limpios_raj)
Grafica_20_palabras(tokens_limpios_leonard)
Grafica_20_palabras(tokens_limpios_howard)

```
