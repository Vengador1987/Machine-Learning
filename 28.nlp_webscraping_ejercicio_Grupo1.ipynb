```python
def enlaces(url_base):
    lista_urls = []
    
    #Obtener el response de la página con Helium sin abrir navegador
    browser = helium.start_chrome(url_base, headless = True)
    sleep(2)

    #Extraemos el html y lo pasamos a objeto bs4
    soup = BeautifulSoup(browser.page_source, "html.parser")
    
    lista_urls = [t.find("a")["href"] for t in soup.find("ul") if t.find("a") != -1][1:]
    
    return lista_urls
```

```python
def dialogos_capitulos(lista_urls):

    personajes = ["Sheldon", "Leonard", "Penny", "Raj", "Howard"]

    df_sheldon = pd.DataFrame(columns = ["temporada", "episodio", "personaje", "dialogo"])
    df_leonard = pd.DataFrame(columns = ["temporada", "episodio", "personaje", "dialogo"])
    df_penny =  pd.DataFrame(columns = ["temporada", "episodio", "personaje", "dialogo"])
    df_raj = pd.DataFrame(columns = ["temporada", "episodio", "personaje", "dialogo"])
    df_howard =  pd.DataFrame(columns = ["temporada", "episodio", "personaje", "dialogo"])
        
    for url in lista_urls:

        #Obtener el response de la página con Helium sin abrir navegador
        browser = helium.start_chrome(url, headless = True)
        sleep(5)

        #Extraemos el html y lo pasamos a objeto bs4
        soup = BeautifulSoup(browser.page_source, "html.parser")

        #Recorremos el html para extraer los diálogos
        dialogos = list()

        for bs in soup.find_all("p", class_ = "MsoNormal"):

            dialogos.append(bs.text)

        temporada = url.split("-")[1]
        episodio = url.split("-")[3]
        
        for personaje in personajes:
            for dialogo in dialogos:
                if dialogo.startswith(f"{personaje}:"):
                    if personaje == "Sheldon":
                        df_sheldon = df_sheldon.append({"temporada" : temporada, "episodio" : episodio, "personaje" : personaje,
                                                        "dialogo" : dialogo.split(f"{personaje}: ")[1]}, ignore_index=True)
                    elif personaje == "Leonard":
                        df_leonard = df_leonard.append({"temporada" : temporada, "episodio" : episodio, "personaje" : personaje,
                                                        "dialogo" : dialogo.split(f"{personaje}: ")[1]}, ignore_index=True)
                    elif personaje == "Penny":
                        df_penny = df_penny.append({"temporada" : temporada, "episodio" : episodio, "personaje" : personaje,
                                                    "dialogo" : dialogo.split(f"{personaje}: ")[1]}, ignore_index=True)
                    elif personaje == "Raj":
                        df_raj = df_raj.append({"temporada" : temporada, "episodio" : episodio, "personaje" : personaje,
                                                "dialogo" : dialogo.split(f"{personaje}: ")[1]}, ignore_index=True)
                    elif personaje == "Howard":
                        df_howard = df_howard.append({"temporada" : temporada, "episodio" : episodio, "personaje" : personaje,
                                                      "dialogo" : dialogo.split(f"{personaje}: ")[1]}, ignore_index=True)

    return df_sheldon, df_leonard, df_penny, df_raj, df_howard
```

```python
lista_df = [variable[0], variable[1], variable[2], variable[3], variable[4]]

dicc = dict()

for df in lista_df:
    # diccionario de 5 textos
    dicc[df["personaje"].unique()[0]] = " ".join(df["dialogo"])

dicc
```

```python
# 4

import nltk

dicc2 = dict()

for personaje, dialogo in dicc.items():
    tokens = nltk.word_tokenize(text = dialogo, language = "english")
    dicc2[personaje] = tokens
    
dicc2   
    
```

```python
# 5
stopwords = nltk.corpus.stopwords.words("english")
type(stopwords)
```

```python
def limpiar_tokens(lista, stopwords):
    tokens_limpios = list()

    tokens = [token.lower() for token in lista]

    for token in tokens: 
        if token not in stopwords:  
            if len(token) > 2:
                tokens_limpios.append(token)
    
    return tokens_limpios
```

```python
for personaje, tokens in dicc2.items():
    dicc_aux = dict()
    dicc_aux["tokens_limpios"] = limpiar_tokens(tokens, stopwords)    
    dicc2[personaje] = dicc_aux

dicc2
```

```python
# 6

def riqueza_lexica(lista):
    total_palabras = len(lista)
    palabras_diferentes = len(set(lista))

    riqueza_lexica = palabras_diferentes / total_palabras
    
    return riqueza_lexica
```

```python
for personaje, dicc_aux in dicc2.items():
    dicc_aux["riqueza_lexica"] = riqueza_lexica(tokens)
    
    dicc2[personaje] = dicc_aux

dicc2
```

```python
# 7
def palabras_frecuentes(lista):
    palabras = list()
    
    texto_nltk = nltk.Text(tokens = lista)
    
    distribucion = nltk.FreqDist(samples = texto_nltk)
    
    palabras = distribucion.most_common(3)
    
    return palabras
```

```python
for personaje, dicc_aux in dicc2.items():

    dicc_aux["palabras_frecuentes"] = palabras_frecuentes(dicc_aux["tokens_limpios"])
    dicc2[personaje] = dicc_aux

dicc2
```

```python
# 8
def palabras_similares(palabra, lista):
    idx = nltk.text.ContextIndex(tokens = lista)

    palabras_similares = idx.similar_words(word = palabra)

    return palabras_similares
```

```python
for personaje, dicc_aux in dicc2.items():
    palabra = dicc_aux["palabras_frecuentes"][0][0]
    dicc_aux["palabras_similares"] = palabras_similares(palabra, dicc_aux["tokens_limpios"])
    dicc2[personaje] = dicc_aux
```

```python
# 9

for personaje, dicc_aux in dicc2.items():
    texto_nltk = nltk.Text(tokens = dicc_aux["tokens_limpios"])
    
    distribucion = nltk.FreqDist(samples = texto_nltk)

    dicc_aux["hapaxes"] = distribucion.hapaxes()
    
    dicc2[personaje] = dicc_aux
```

```python
# 10

import matplotlib.pyplot as plt

for personaje, dicc_aux in dicc2.items():
    print(f"Personaje: {personaje}")
    nltk.FreqDist(nltk.Text(dicc_aux["tokens_limpios"])).plot(20)
    plt.show()
```
