## Ejercicio OpenCV: Dibujar sobre imagenes.

Usando la imagen **`sistema_solar.jpg`** realiza los siguiente ejercicios.

### Ejercicio 1:
- Lee el archivo usando **`OpenCV`**.

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# OpenCV
import cv2
```

```python
img = cv2.imread(filename = "sistema_solar.jpg")

type(img)
```

```python
img.shape
```

```python
plt.imshow(img)
plt.show()
```

### Ejercicio 2:
- Gira la imagen sobre el eje x.

```python
v_img = cv2.flip(src = img, flipCode = 0)

plt.imshow(v_img)
plt.show()
```

### Ejercicio 3:
- Dibuja un circulo rojo vacio sobre el planeta tierra.

```python
plt.imshow(img)
plt.show()
```

```python
cv2.circle(img = img, center = (355, 259), radius = 45, color = (255, 0, 0), thickness = 5)

plt.imshow(img)
plt.show()
```

### Ejercicio 4:
- Dibuja un poligono de 3 esquinas (un triangulo) que toque 3 planetas diferentes que no este vacio.

```python
vertices = np.array([[255, 525],
                     [150, 300],
                     [350, 250]], np.int32)

pts = vertices.reshape((-1, 1, 2))

pts
```

```python
pts.shape
```

```python
cv2.fillPoly(img = img, pts = [pts], color = (255, 0, 0))

plt.imshow(img)
plt.show()
```
